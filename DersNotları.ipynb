# **DERS 1**
BİST100 Analizi

```python
import yfinance as yf
import plotly.express as px
```

```python
B100=yf.download('XU100.IS',start="2022-01-01",end="2022-12-31")
B100
```

```python
B100.head(10)
```

```python
B100.info()
```

```python
B100.describe()
```

```python
#satır silme
B100=B100[B100.index>"2022-01-01"]
B100
```

```python
#sütun silme
B100=B100[['Adj Close']]
B100
```

```python
#isim değiştirme
B100.columns=["kapanış(try)"]
B100
```

```python
norm=B100.iloc[0,0]
B100["Normalize"]=B100["kapanış(try)"]/norm
B100
```

# *BIST 100 Endeksi 2022 yılı başından beri yaklaşık 2,9 katına çıktı!*

```python
B100["normalize"].plot()
```

```python
fig1=px.line(B100, y="Normalize")
fig2=px.bar(B100, y="Normalize")
fig3=px.scatter(B100, y="Normalize")
fig1.show()
fig2.show()
fig3.show()
```

```python
from plotly.graph_objs import XAxis
px.line(B100, y="Normalize",
        width=1000, height=500,
        title= "BIST100 Endeksi 2022 - Normalize (1 Ocak 2022)",
        hover_data={"Normalize":".2f", "kapanış(try)":".2f"})
```
