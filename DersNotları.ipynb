# **DERS 2**

```python
#kur veri çekimi
Kur=yf.download('TRY=X', start='2022-01-01', end="2022-12-31")
Kur=Kur[Kur.index>"2022-01-01"]
Kur=Kur[['Adj Close']]
Kur.columns=["kapanış(try)"]
Kur["Normalize USD/TRY"]=Kur["kapanış(try)"]/Kur.iloc[0,0]
```

```python
B100K=B100.merge(Kur, left_index=True, right_index=True)
B100K
```

```python
px.line(B100K, y=["Normalize","Normalize USD/TRY"])
```

```python
fig=go.Figure()
fig.add_trace(go.Bar(x=B100K.index,y=B100K["Normalize"], name="Normalize BIST100"))
fig.add_trace(go.Scatter(x=B100K.index,y=B100K["Normalize USD/TRY"], name="Normalize USD/TRY"))
fig.update_layout(title="Normalize BIST100 ve Normalize Kur",
                  xaxis_title="Tarih", xaxis_dtick="M1", yaxis_title="kapanış(try)")
```

```python
from pandas.core.series import Axis
indirme_listesi=['XU100.IS','TRY=X']
Veri_2023=pd.DataFrame()

for i in indirme_listesi:
  data = yf.download(i, start='2023-01-01', end='2023-04-01')
  data=data["Adj Close"].rename(i)
  Veri_2023=pd.concat([Veri_2023, data], axis=1)

Veri_2023
```

```python
Veri_2023.columns=["kapanış(try)","USD/TRY"]
Veri_2023=Veri_2023[Veri_2023["kapanış(try)"]>0]
Veri_2023
```

```python
Veri_2022 = B100K[["kapanış(try)","USD/TRY"]]
Veri_2022
```

```python
Veri=pd.concat([Veri_2022, Veri_2023], axis=0)
Veri
```

```python
Veri['Normalize']=Veri["kapanış(try)"]/Veri.iloc[0,0]
# Veri['Normalize USD/TRY'] = Veri['USD/TRY']/Veri.iloc[0,1]
Veri['Normalize USD/TRY'] = Veri['USD/TRY']/Veri.loc[Veri.index[0],"USD/TRY"]
Veri['degisim']=Veri['Normalize'].diff(1)
Veri['d_yon']=Veri['degisim'].map(lambda x: 'düsen gün' if x <0 else 'yükselen gün')
Veri['degisim%'] = Veri['degisim']/Veri['Normalize'].shift(1)*100
Veri
```

```python
fig=go.Figure()
fig.add_trace(go.Bar(x=Veri.index,y=Veri["Normalize"], name="Normalize BIST100"))
fig.add_trace(go.Scatter(x=Veri.index,y=B100K["Normalize USD/TRY"], name="Normalize USD/TRY"))
fig.update_layout(title="Normalize BIST100 ve Normalize Kur",
                  xaxis_title="Tarih", xaxis_dtick="M1", yaxis_title="kapanış(try)")
```
